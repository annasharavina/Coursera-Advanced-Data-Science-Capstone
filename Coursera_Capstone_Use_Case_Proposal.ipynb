### Use Case: Airplane Accident Severity Prediction
### Problem Statement: 
For years, air travel has been the preferred means of transportation due to its time efficiency, cost-effectiveness, and high level of convenience. According to the FAA, as of June 2019, a staggering 2,781,971 passengers fly daily in the US. Passengers generally perceive flying as safe, given the rigorous inspections and security measures in place to prevent and address potential incidents. Nevertheless, there remains a slim possibility of unfortunate occurrences.

Consider being tasked with a project by a leading airline. Your objective is to develop machine learning models that can forecast and categorize the severity of airplane accidents based on historical incidents. By accomplishing this, all airlines, and indeed the entire aviation industry, would be able to foresee the potential severity of airplane accidents caused by various factors and subsequently formulate action plans to minimize associated risks.
